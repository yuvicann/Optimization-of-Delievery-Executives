```python
def calculate_distance(lat1, lon1, lat2, lon2):
    coords_1 = (lat1, lon1)
    coords_2 = (lat2, lon2)
    return geopy.distance.geodesic(coords_1, coords_2).kilometers
```

```python
soln = pd.DataFrame( columns=['de_id', 'order_id','cost','time','dist_km','time_promise'])
```

```python
cost_per_km = 4  # Cost per kilometer
cost_per_min = 1  # Cost per min for overall
speed_of_travel = 20  # Speed of travel in km/h

for executive_index, executive_row in df_executives.iterrows():
    for order_index, order_row in df_orders.iterrows():
        distance1 = calculate_distance(executive_row['de_latitude'], executive_row['de_longitude'],
                                      order_row['store_latitude'], order_row['store_longitude'])
        packing_time = order_row['ord_packing_time']
        time_to_free=executive_row['de_time_to_free']
        distance2= order_row['distance']
        distance=distance1+distance2
        travel_time=(distance/speed_of_travel)*60
        
        time= packing_time+time_to_free+travel_time
        cost = (distance * cost_per_km) + (time * cost_per_min)
        time_promise=order_row['time_promise']
        new_row  = {"de_id": executive_row["de_id"], "order_id": order_row['order_id'], "cost":cost, 
                    "time":time, "dist_km":distance, "time_promise":time_promise  }
        if time<=time_promise:
            soln=soln._append(new_row, ignore_index = True)
```

```python
soln.head(25)
```

```python
df_within_time_promise = soln[soln['time'] <= soln['time_promise']]
```

```python
# Group by order ID and select the delivery with the minimum cost
best_suited_deliveries = df_within_time_promise.loc[df_within_time_promise.groupby('order_id')['cost'].idxmin()]
```

```python
# Reset index to remove multi-level index
best_suited_deliveries.reset_index(drop=True, inplace=True)

best_suited_deliveries.head(25)
```

```python

# Plotting
plt.figure(figsize=(15, 10))
plt.plot(best_suited_deliveries.index, best_suited_deliveries['time'], marker='*', label='Time of Delivery')
plt.plot(best_suited_deliveries.index, best_suited_deliveries['time_promise'], marker='+', label='Time Promise')
plt.xlabel('Order')
plt.ylabel('Time')
plt.title('Time of Delivery vs Time Promise')
plt.legend()
plt.grid(True)
plt.xticks(best_suited_deliveries.index, best_suited_deliveries['order_id'])
plt.tight_layout()
plt.show()
```

```python
# Reset index to remove multi-level index
best_suited_deliveries.reset_index(drop=True, inplace=True)

best_suited_deliveries["de_id"] = best_suited_deliveries["de_id"].astype(int)

best_suited_deliveries["order_id"] = best_suited_deliveries["order_id"].astype(int)

best_suited_deliveries.to_excel("final_output.xlsx")
```
