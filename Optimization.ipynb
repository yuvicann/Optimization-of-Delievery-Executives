# Assumptions
## cost_per_km = 4 , Cost per kilometer
## cost_per_min = 1 , Cost per min for overall
## speed_of_travel = 20 , Speed of travel in km/h

```python
def calculate_distance(lat1, lon1, lat2, lon2):
    coords_1 = (lat1, lon1)
    coords_2 = (lat2, lon2)
    return geopy.distance.geodesic(coords_1, coords_2).kilometers
```

```python
final_soln = pd.DataFrame( columns=['de_id', 'order_id','cost','time','dist_km','time_promise'])
```

```python
cost_per_km = 4  # Cost per kilometer
cost_per_min = 1  # Cost per min for overall
speed_of_travel = 20  # Speed of travel in km/h



for executive_index, executive_row in df_executives.iterrows():
    for order_index, order_row in df_orders.iterrows():

        
        distance1 = calculate_distance(executive_row['de_latitude'], executive_row['de_longitude'],
                                      order_row['store_latitude'], order_row['store_longitude'])
        packing_time = order_row['ord_packing_time']
        time_to_free = executive_row['de_time_to_free']
        distance2 = order_row['distance']
        distance = distance1+distance2


        
        travel_time=(distance/speed_of_travel)*60


        
        
        time= packing_time+time_to_free+travel_time
        cost = (distance * cost_per_km) + (time * cost_per_min)
        time_promise=order_row['time_promise']
        new_row  = {"de_id": executive_row["de_id"], "order_id": order_row['order_id'], "cost":cost, 
                    "time":time, "dist_km":distance, "time_promise":time_promise  }



        
        if time<=time_promise:
            soln=soln._append(new_row, ignore_index = True)
```

```python
soln.info()
```

```python
df_within_time_promise = soln[soln['time'] <= soln['time_promise']]
```

#### Here we are optimising the total time taken should be less than the promised time.

```python
df_within_time_promise.info()
```

#### Here we can se that there are about 95k entries, and there are many delievery executives that can order this within the promise time, now we have to find the orders with minimum cost and minimum time.

```python
# Grouping by order ID and selecting the delivery with the minimum cost

best_suited_deliveries = df_within_time_promise.loc[df_within_time_promise.groupby('order_id')['cost'].idxmin()]
```

#### The 'order_id' column serves as a unique identifier for each order, and the 'cost' column represents the estimated delivery cost for a particular order-delivery executive pairing.

#### Within each group (i.e., for each unique order ID), it finds the index of the row that has the minimum value in the 'cost' column using the idxmin() method.

#### After that, we are using .loc to retrieve the rows corresponding to those minimum cost indices, effectively selecting the most cost-effective delivery option for each order.



#### Now we have optimised the time as well as cost of the delievery that will be assigned to the particular order id.

```python
# Resetting index to remove multi-level index


best_suited_deliveries.reset_index(drop=True, inplace=True)

```

```python
import matplotlib.pyplot as plt
plt.scatter(best_suited_deliveries['time_promise'], best_suited_deliveries['time'],color="black")
plt.xlabel("Promise Time (minutes)")
plt.ylabel("Actual Time (minutes)")
plt.title("Promise Time vs. Actual Time")
plt.show()
```

## Interpretation:
#### Points above the 45-degree line: Delivery took longer than promised.
#### Points below the line: Delivery was faster than promised.
#### Tight clustering around the line: Good consistency 

### As we can see most of the orders are below 45 degree line, hence delievery was faster.
